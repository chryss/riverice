```python
import pandas as pd
from pathlib import Path
import datetime as dt
import seaborn as sns
```

```python
fiwepsdir = Path("/Volumes/CWMobileSSD/ABoVE_WAK/")
datadir = Path("../data")
```

```python
cjcpath = fiwepsdir / "predictors_DF/teleconnections.csv"
predpath = datadir / "predictors/teleconnections_20240320.csv"
predictor_df = pd.read_csv(predpath)
predictor_df = predictor_df[predictor_df.Year.between(1980, 2023) ].reset_index().dropna(axis=1)
predictor_df.drop(columns=['index'], inplace=True)
predictor_df.set_index("Year", inplace=True)
predictor_df
```

```python
breakupdata = Path("../data/breakupdata/derived/breakupDate_cleaned.csv")
breakup = pd.read_csv(breakupdata, skiprows=3, index_col=0)
breakup
```

```python
breakup_sum = breakup[['siteID', 'JulianDay']].groupby(
    "siteID").mean().sort_values(by="JulianDay")
breakup_sum['DateMean'] = (
    pd.to_timedelta(breakup_sum.JulianDay, unit='D') + pd.to_datetime("2024-01-01")
    ).dt.strftime("%Y-%m-%d")
breakup_sum
```
